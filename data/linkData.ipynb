```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2_LinkData_2024110815025924_100.csv')
df.head()
```

```python
len(df["ID"].unique())
```

```python
id_number = 319000330
filtered_df = df[df['ID'] == id_number]

avgflux = filtered_df['avgAverageSpeed'] * filtered_df['avgDensity']

plt.figure(figsize=(10, 6))
plt.plot(filtered_df['SimTime0'], avgflux, marker='o')
plt.title(f'ID {id_number} ')
plt.xlabel('SimTime0')
plt.ylabel('avgFlux')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# 绘制散点图，颜色根据Time Step渐变
scatter = plt.scatter(filtered_df['avgDensity'], avgflux, c=filtered_df['SimTime0'], cmap='viridis', s=5)
# 添加颜色条
cbar = plt.colorbar(scatter)
cbar.set_label('Time Step')
plt.title(f'ID {id_number} ')
plt.xlabel('avgDensity')
plt.ylabel('avgFlux')
plt.grid(True)
plt.show()
```
