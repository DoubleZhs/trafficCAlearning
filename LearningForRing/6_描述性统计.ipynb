```python
num_DiDi_lst = [10,20,40,60,80,100,150,200]
pro_Autonavi_lst = [0.001,0.005,0.01] + list(np.arange(0,0.31,0.05)[1:])
pro_Autonavi_lst = [round(i,3) for i in pro_Autonavi_lst]
```

```python
x = np.arange(len(num_DiDi_lst))
y1 = [] # 0-10天
y2 = [] # 11-20天
y3 = [] # 21-30天

width = 0.2
for num_DiDi in num_DiDi_lst:
    df = pd.read_csv(f"Feature/train/VehicleData_n{num_DiDi}.csv",encoding='utf-8')
    
    for date in [0,10,20]:
        sub_df = df[df['Date'].between(date,date+9)]
        num = sub_df[sub_df['ClosedVehicle']==True].shape[0]
        if date == 0 : y1.append(num)
        elif date == 10 : y2.append(num)
        else : y3.append(num) 

fig, ax = plt.subplots(dpi=300, figsize=(12, 4))
# Create grouped bars
bar1 = ax.bar(x - width, y1, width, label='1-10 day')
bar2 = ax.bar(x , y2, width, label='11-20 day')
bar3 = ax.bar(x + width, y3, width, label='21-30 day')
# Add labels, title, and legend
ax.set_xlabel('num_DiDi')
ax.set_ylabel('num of trips')
# ax.set_title('Grouped Bar Chart')
ax.set_xticks(x)
ax.set_xticklabels([str(xx) for xx in num_DiDi_lst])
ax.legend()
plt.show()

```

```python
x = np.arange(len(pro_Autonavi_lst))
y1 = [] # 0-10天
y2 = [] # 11-20天
y3 = [] # 21-30天

width = 0.2
for pro in pro_Autonavi_lst:
    df = pd.read_csv(f"Feature/train/VehicleData_n10.csv",encoding='utf-8')
    
    for date in [0,10,20]:
        sub_df = df[df['Date'].between(date,date+9)]
        num = sub_df[(sub_df['ClosedVehicle']==False)&(sub_df['Tag']<=pro)].shape[0]
        if date == 0 : y1.append(num)
        elif date == 10 : y2.append(num)
        else : y3.append(num) 

fig, ax = plt.subplots(dpi=300, figsize=(12, 4))
# Create grouped bars
bar1 = ax.bar(x - width, y1, width, label='1-10 day')
bar2 = ax.bar(x , y2, width, label='11-20 day')
bar3 = ax.bar(x + width, y3, width, label='21-30 day')
# Add labels, title, and legend
ax.set_xlabel('pro_Autonavi')
ax.set_ylabel('num of trips')
# ax.set_title('Grouped Bar Chart')
ax.set_xticks(x)
ax.set_xticklabels([str(xx) for xx in pro_Autonavi_lst])
ax.legend()
plt.show()

```
