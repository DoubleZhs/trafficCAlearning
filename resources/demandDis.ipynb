```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure inline plotting for Jupyter notebooks
%matplotlib inline
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('DemandTimeDistribution.csv')

# Display the first few rows of the DataFrame to ensure it loaded correctly
df.head()
```

```python
def convert_seconds_to_time(seconds):
    hours = (seconds // 2400) % 24
    minutes = (seconds % 2400) // 40
    return f"{hours:02}:{minutes:02}"

tranT = [convert_seconds_to_time(t) for t in df['T']]

plt.figure(figsize=(10, 6))
plt.plot(tranT, df['Pro'])
plt.title('Distribution of Pro over T')
plt.xlabel('Time')
plt.ylabel('Pro')
plt.grid(True)

hours = [f"{h:02}:00" for h in range(24)]
plt.xticks(hours)

plt.show()
```
